```python
#Login system with 3 attempts
class LoginError(Exception):
    pass
correct_username="nandh"
correct_password="1005"
attempts=3
try:
    while attempts>0:
        username=input("Enter username:")
        password=input("Enter password:")

        if username==correct_username and password==correct_password:
            print("Login Successfull!")
            break
        else:
            attempts-=1
            print(f"Wrong credentials!\n {attempts} Attempts left")
    if attempts==0:
        raise LoginError("Too many wrong attempts. Access Denied!")
except LoginError as e:
     print(e)
except ValueError:
     print("Invalid input!")
```

```python
#ATM 
withdraw=15.00
while withdraw>0:
    amount=float(input("Enter amount to withdraw:"))
    balance=withdraw-amount
    if amount<=withdraw:
        print(f"Your withdrawal completed with Rs.{amount}")
        print(f"Your current balance is {balance}.")
        break
    else:
        print(f"Your bank balance is insufficient.Kindly check your balance.")
        break



#using Try Except:
balance=10000
try:
    amount=float(input("Enter amount to withdraw:"))
    if amount>balance:
        print("Insufficient Balance!")
    else:
        balance-=amount
        print("Your withdrawal successfull.")
        print("Balance amount:",balance)
except ValueError:
    print("Enter valid value!")
except InsufficientBalancerError as e:
    print(e)
```

```python
#Custom error with agelimit
class AgeError(Exception):
    pass

try:
    age=int(input('Enter your age:'))
    if age>60:
        raise AgeError("You're too Old!")
    elif age<18:
        raise AgeError("You're too Young!")
    else:
        print("You're Eligible")

except AgeError as e:
    print("CustomError:" , e)
except ValueError:
    print("Enter valid value!")
```

```python
#Multiple Exceptions
try:
    num1=input("Enter numerator:")
    num2=input("Enter denominator:")
    result=(num1/num2)
    print(result)
except ZeroDivisionError:
    print("The numerator can't divide by zero!")
except ValueError:
    print("Enter a valid value!")
except TypeError:
    print("Cannot mix different data types!")
```

```python
#Using finally
try:
    num1=float(input("Enter numerator:"))
    num2=float(input("Enter denominator:"))
    result=num1/num2
except ValueError:
    print("Enter valid input!")
except ZeroDivisionError:
    print("Not divisible by zero!")
else:
    print(result)
finally:
    print("Program Ended.")
    
```

```python
#Update records
file3=open("students.txt","r")
lines=file3.readlines()
file3.close()
updated_lines=[]
for line in lines:
    data=line.strip().split(",")
    name=data[0]
    if name=="Roger":
        new_line="Rog,100,100,100\n"
        updated_lines.append(new_line)
    else:
        updated_lines.append(line)

file3=open("students.txt","w")
file3.writelines(updated_lines)
file3.close()
print("Records Updated")
```
